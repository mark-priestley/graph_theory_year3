# Graph theory: Modelling college courses

csv files already generated from cypher queries

This notebook models how many students would change courses, and what courses would they change to.

Results fully explained in "final_results.pdf"

```python
import pandas as pd
import numpy as np
import json

df = pd.read_csv("less_than_2.csv")
df2 = pd.read_csv("less_than_15.csv")
df3 = pd.read_csv("less_than_25.csv")
```

```python
df3.head()
```

```python
courses = list(df.loc[:,"source"])

# Inclusive
d = {}

i = 0
while i < len(courses):
  if courses[i] not in d.keys():
    if i != 0:
      d[courses[i - 1]].append(i - 1)
    d[courses[i]] = [i]
  i += 1

d[courses[i - 1]].append(len(courses) - 1)

courses = list(df2.loc[:,"source"])

# Inclusive
d2 = {}

i = 0
while i < len(courses):
  if courses[i] not in d2.keys():
    if i != 0:
      d2[courses[i - 1]].append(i - 1)
    d2[courses[i]] = [i]
  i += 1

d2[courses[i - 1]].append(len(courses) - 1)

courses = list(df3.loc[:,"source"])

# Inclusive
d3 = {}

i = 0
while i < len(courses):
  if courses[i] not in d3.keys():
    if i != 0:
      d3[courses[i - 1]].append(i - 1)
    d3[courses[i]] = [i]
  i += 1

d3[courses[i - 1]].append(len(courses) - 1)
```

```python
df_v_3 = pd.DataFrame(columns=["Course","Possible Courses","Number of Possible Courses", "Average Distance to Possible Courses"])

for course in d3.keys():
  start, end = d3[course][0], d3[course][1]
  a = df3.loc[start:end,"target"]
  b = df3.loc[start:end, "distance"]
  tmp = pd.DataFrame(data = [[course, ", ".join(list(a)), len(a), b.mean()]],columns=["Course","Possible Courses","Number of Possible Courses", "Average Distance to Possible Courses"])
  df_v_3 = df_v_3.append(tmp,ignore_index=True)
```

```python
df_v_2 = pd.DataFrame(columns=["Course","Possible Courses","Number of Possible Courses", "Average Distance to Possible Courses"])

for course in d2.keys():
  start, end = d2[course][0], d2[course][1]
  a = df2.loc[start:end,"target"]
  b = df2.loc[start:end, "distance"]
  tmp = pd.DataFrame(data = [[course, ", ".join(list(a)), len(a), b.mean()]],columns=["Course","Possible Courses","Number of Possible Courses", "Average Distance to Possible Courses"])
  df_v_2 = df_v_2.append(tmp,ignore_index=True)
```

```python
df_v_2['Average Distance to Possible Courses'].mean()
```

```python
# Adding students in each course

df_v_2["Start Students"] = [30,50,20,30,100,40,120,30,20,60,50,30,50,30,80,30,120,50,40] 
```

```python
# 10% of students leave

df_v_2["Outgoing Students"] = (df_v_2["Start Students"] * 0.1).astype(int)
df_v_2["Incoming Students"] = 0
```

```python
import random

# Reset Incoming Students
df_v_2["Incoming Students"] = 0

outgoing_courses = {}

for index, row in df_v_2.iterrows():
  outgoing = row['Outgoing Students']
  outgoing_courses[row['Course']] = {}

  start, end = d2[row['Course']][0], d2[row['Course']][1]
  possible_courses = list(df2.loc[start:end,"target"])
  for course in possible_courses:
    outgoing_courses[row['Course']][course] = 0

  i = 0
  while i < outgoing:
    student_new_course = random.choice(possible_courses)
    df_v_2.loc[df_v_2["Course"] == student_new_course,"Incoming Students"] = df_v_2.loc[df_v_2["Course"] == student_new_course,"Incoming Students"] + 1
    outgoing_courses[row['Course']][student_new_course] += 1
    i += 1

df_v_2['End Students'] = df_v_2['Start Students'] - df_v_2['Outgoing Students'] + df_v_2['Incoming Students']
```

```python
# Write outgoing students file
with open("outgoing_students.json", "w") as fd:
    json.dump(outgoing_courses, fd, indent=3)
```
